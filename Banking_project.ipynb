```python
import matplotlib.pyplot as plt  # Import matplotlib for plotting
import seaborn as sns  # Import seaborn for statistical plotting

# Loop through each predictor column and create count plots
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards", "Nationality", "Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)  # Create a new figure for each plot
    sns.countplot(data=df, x=predictor)  # Create count plot with GenderId as hue
    plt.title(f'Count Plot for {predictor}')  # Add title for clarity
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to prevent overlapping
    plt.show()  # Display the plot
```

```python
### Histplot of value count for fifferent Occupations
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
### Numerical Analysis
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts', 'Foreign Currency Account', 'Saving Accounts','Business Lending',]

#Univarient Analysis and Visualization
plt.figure(figsize=(20,17))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

```python
###Heat maps
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts', 'Foreign Currency Account', 'Saving Accounts','Business Lending',]
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot = True, cmap='crest',fmt = ".2f")
plt.title("correlation_matrix")
plt.show()
```
